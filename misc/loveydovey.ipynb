```python
class LoveyDovey:
    def __init__(self, boy, girl):
        self.boy = boy
        self.girl = girl
        self.loveydovey = self._boy_love_girl()
        self.doveylovey = self._girl_love_boy()
    def _boy_love_girl(self):
        return self.boy+' lovey '+self.girl
    def _girl_love_boy(self):
        return self.girl+' lovey '+self.boy

```

```python
juju = LoveyDovey('Mummy', 'Bubbu')
```

```python
juju.loveydovey
```

```python
juju.doveylovey
```
